2. You are analyzing sports teams. Members of each team are stored in a list. The Coach is the first name in the list, the captain is the second name in the list, and other players are listed after that. These lists are stored in another list, which starts with the best team and proceeds through the list to the worst team last. Complete the function below to select the captain of the worst team.

````python
def losing_team_captain(teams):
    """Diberikan daftar tim, di mana setiap tim adalah daftar nama, kembalikan pemain kedua (kapten)
    dari tim terakhir yang terdaftar.
    """
    if len(teams) == 0:
        return None  # Jika daftar tim kosong, kembalikan None
    else:
        last_team = teams[-1]  # Tim terakhir dalam daftar
        if len(last_team) >= 2:
            return last_team[1]  # Kapten adalah pemain kedua dalam tim terakhir
        else:
            return None  # Jika tim terakhir memiliki kurang dari 2 pemain, kembalikan None

# Contoh penggunaan:
teams = [['Tim A', 'Pemain 1', 'Pemain 2'], ['Tim B', 'Kapten B'], ['Tim C']]

# Panggil fungsi losing_team_captain
kapten = losing_team_captain(teams)

# Cetak hasil
if kapten is not None:
    print(f"Kapten dari tim terakhir: {kapten}")
else:
    print("Tidak ada tim atau tim terakhir tidak memiliki kapten")
````

Implement a function that reproduces this behaviour, returning a list of booleans corresponding to whether the corresponding element is greater than n.

````python
def elementwise_greater_than(L, thresh):
    """Mengembalikan daftar dengan panjang yang sama seperti L, di mana nilai pada indeks i adalah
    True jika L[i] lebih besar dari thresh, dan False sebaliknya.
    """
    # Inisialisasi daftar hasil dengan daftar kosong
    result = []

    # Perulangan melalui setiap elemen dalam daftar L
    for num in L:
        # Periksa jika elemen lebih besar dari thresh
        if num > thresh:
            result.append(True)
        else:
            result.append(False)

    return result

# Contoh penggunaan:
L = [1, 2, 3, 4]
thresh = 2
hasil = elementwise_greater_than(L, thresh)
print(hasil)
````

On average, how much money can you expect to gain (or lose) every time you play the machine? The casino keeps it a secret, but you can estimate the average value of each pull using a technique called the Monte Carlo method. To estimate the average outcome, we simulate the scenario many times, and return the average result.

Complete the following function to calculate the average value per play of the slot machine.

````python
import random

def estimate_average_slot_payout(n_runs):
    """Menjalankan mesin slot sebanyak n_runs kali dan mengembalikan rata-rata keuntungan bersih per putaran.
    """
    total_profit = 0  # Inisialisasi total keuntungan bersih
    for _ in range(n_runs):
        # Simulasi satu putaran mesin slot
        payout = random.randint(0, 100)  # Simulasi pembayaran acak antara 0 hingga 100
        cost = 1  # Biaya satu putaran mesin slot
        net_profit = payout - cost  # Keuntungan bersih dari satu putaran
        total_profit += net_profit  # Tambahkan keuntungan bersih ke total

    # Hitung rata-rata keuntungan bersih per putaran
    average_profit_per_run = total_profit / n_runs
    return average_profit_per_run

# Contoh penggunaan:
n_runs = 1000  # Anda dapat mengubah jumlah putaran mesin slot sesuai keinginan
rata_rata_keuntungan = estimate_average_slot_payout(n_runs)
print(f"Rata-rata Keuntungan Bersih per Putaran: {rata_rata_keuntungan}")
````

````python
# Fungsi untuk mencari kata kunci dalam satu dokumen
def word_search(doc_list, keyword):
    # List untuk menyimpan indeks dokumen yang sesuai
    indices = []

    # Melakukan iterasi melalui indeks (i) dan elemen (doc) dari dokumen
    for i, doc in enumerate(doc_list):
        # Memisahkan string doc menjadi daftar kata (berdasarkan spasi)
        tokens = doc.split()

        # Membuat daftar tertransformasi di mana setiap kata 'dinormalisasi' untuk memfasilitasi pencocokan.
        # Titik dan koma dihapus dari akhir setiap kata, dan kata tersebut diubah menjadi huruf kecil.
        normalized = [token.rstrip('.,').lower() for token in tokens]

        # Apakah ada yang cocok? Jika ya, perbarui daftar indeks yang cocok.
        if keyword.lower() in normalized:
            indices.append(i)

    # Mengembalikan daftar indeks yang sesuai
    return indices

# Fungsi untuk mencari beberapa kata kunci dalam beberapa dokumen
def multi_word_search(doc_list, keywords):
    # Kamus untuk menyimpan hasil pencarian kata kunci
    keyword_to_indices = {}

    # Melakukan pencarian untuk setiap kata kunci dalam daftar kata kunci
    for keyword in keywords:
        # Menggunakan fungsi word_search untuk mencari kata kunci dalam dokumen
        hasil_pencarian = word_search(doc_list, keyword)

        # Menyimpan hasil pencarian dalam kamus dengan kata kunci sebagai kunci (key) kamusnya
        keyword_to_indices[keyword] = hasil_pencarian

    # Mengembalikan kamus hasil pencarian
    return keyword_to_indices

# Contoh penggunaan:
doc_list = ["Ini adalah dokumen pertama.", "Ini adalah dokumen kedua.", "Ini adalah dokumen ketiga."]
keywords = ["kedua", "pertama"]
hasil_pencarian = multi_word_search(doc_list, keywords)
print(hasil_pencarian)

````

Suppose we wanted to create a new type to represent hands in blackjack. One thing we might want to do with this type is overload the comparison operators like >n ad <= so that we could use them to check whether one hand beats another. e.g. it'd be cool if we could do this:

>>> hand1 = BlackjackHand(['K', 'A'])
>>> hand2 = BlackjackHand(['7', '10', 'A'])

True
Well, we're not going to do all that in this question (defining custom classes is a bit beyond the scope of these lessons), but the code we're asking you to write in the function below is very similar to what we'd have to write if we were defining our own BlackjackHand class. (We'd put it in the __gt__ magic method to define our custom behaviour for >.)

Fill in the body of the blackjack_hand_greater_than function according to the docstring.


````python
def blackjack_hand_greater_than(hand_1, hand_2):
    # Fungsi ini menghitung total nilai tangan berdasarkan aturan blackjack.
    def calculate_hand_total(hand):
        total = 0  # Total nilai awal tangan.
        num_aces = 0  # Jumlah kartu As dalam tangan.

        for card in hand:
            if card.isdigit():
                total += int(card)  # Kartu bernilai angka, tambahkan nilainya.
            elif card in ['K', 'Q', 'J']:
                total += 10  # Kartu wajah (Raja, Ratna, Jack) bernilai 10.
            elif card == 'A':
                num_aces += 1  # Hitung kartu As.
                total += 11  # Asumsikan nilai kartu As sebagai 11.

        # Periksa apakah ada kartu As yang harus dihitung sebagai 1 untuk menghindari melebihi 21.
        while num_aces > 0 and total > 21:
            total -= 10  # Ubah nilai kartu As dari 11 menjadi 1.
            num_aces -= 1

        return total  # Kembalikan total nilai tangan.

    total_1 = calculate_hand_total(hand_1)  # Hitung total nilai tangan pertama.
    total_2 = calculate_hand_total(hand_2)  # Hitung total nilai tangan kedua.

    # Tentukan hasil permainan berdasarkan aturan blackjack.
    return total_1 <= 21 and (total_1 > total_2 or total_2 > 21)
````
